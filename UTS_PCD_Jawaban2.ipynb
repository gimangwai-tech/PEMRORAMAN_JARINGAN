```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Baca gambar (gunakan gambar grayscale)
img = cv2.imread('noise.png', cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError("Gambar 'noise.png' tidak ditemukan!")

# 2. Definisikan tiga kernel
LPF = (1/9) * np.array([[1,1,1],
                        [1,1,1],
                        [1,1,1]])  # Low Pass Filter

HPF = np.array([[-1,-1,-1],
                [-1, 8,-1],
                [-1,-1,-1]])     # High Pass Filter

BPF = np.array([[ 0,-1, 0],
                [-1, 5,-1],
                [ 0,-1, 0]])     # Band Pass Filter

# 3. Terapkan filter menggunakan konvolusi (cv2.filter2D)
lpf_img = cv2.filter2D(img, -1, LPF)
hpf_img = cv2.filter2D(img, -1, HPF)
bpf_img = cv2.filter2D(img, -1, BPF)

# 4. Tampilkan hasil dalam satu figure
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(lpf_img, cmap='gray')
plt.title('Low Pass Filter (Blur)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(hpf_img, cmap='gray')
plt.title('High Pass Filter (Edge Detection)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(bpf_img, cmap='gray')
plt.title('Band Pass Filter (Sharpening)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
