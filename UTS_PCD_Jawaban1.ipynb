```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar grayscale
img = cv2.imread('bunga.jpg', cv2.IMREAD_GRAYSCALE)

# Pastikan gambar tidak None
if img is None:
    raise FileNotFoundError("File 'bunga.jpg' tidak ditemukan!")

# 1. Citra asli
original = img.copy()

# 2. Citra negatif
negatif = 255 - img

# 3. Transformasi logaritmik (versi diperbaiki)
img_float = img.astype(float)
c = 255 / np.log(1 + np.max(img_float))
log_transform = c * np.log(1 + img_float)
log_transform = cv2.normalize(log_transform, None, 0, 255, cv2.NORM_MINMAX)
log_transform = np.uint8(log_transform)

# 4. Perataan histogram
hist_eq = cv2.equalizeHist(img)

# 5. Tampilkan hasil
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(original, cmap='gray')
plt.title('Citra Asli (Grayscale)')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(negatif, cmap='gray')
plt.title('Citra Negatif')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(log_transform, cmap='gray')
plt.title('Transformasi Logaritmik (Diperbaiki)')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(hist_eq, cmap='gray')
plt.title('Perataan Histogram')
plt.axis('off')

plt.tight_layout()
plt.show()
```
